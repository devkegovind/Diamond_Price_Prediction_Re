```python
df1 = pd.DataFrame.from_dict(attendance_records1, orient='index', columns=range(43))
df1
```

```python
replacement = {'present':'P', 'absent':'A'}

df1 = df1.replace(replacement)
df1
```

```python
import random

students = ['Alice', 'Bob', 'Charlie', 'Dave']
num_days = 30
percent_present = 80

num_present = int(len(students) * percent_present / 100)
present_students = random.sample(students, num_present)

attendance = {student: [] for student in students}

for i in range(num_days):
    for student in students:
        if student in present_students:
            record = 'present'
        else:
            record = 'absent'
        attendance[student].append(record)

for student in students:
    present = attendance[student].count('present')
    absent = attendance[student].count('absent')
    print(f"{student}: present={present}, absent={absent}")
```
