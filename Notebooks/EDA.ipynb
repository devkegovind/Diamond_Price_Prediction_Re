#### **Plots All Numerical Columns**

```python
plt.figure(figsize=(35, 15))
plt.suptitle("Histplots of Numerical Columns")
for i in enumerate(df[['carat', 'depth', 'table', 'x', 'y', 'z', 'price']]):
    plt.subplot(4,2, i[0]+1)
    sns.histplot(df[i[1]], kde = True)
plt.show()
```

```python
plt.figure(figsize=(35, 15))
plt.suptitle("Distplots of Numerical Columns")
for i in enumerate(df[['carat', 'depth', 'table', 'x', 'y', 'z', 'price']]):
    plt.subplot(4,2, i[0]+1)
    sns.distplot(df[i[1]])
plt.show()
```

```python
sns.heatmap(df[['carat', 'depth', 'table', 'x', 'y', 'z', 'price']].corr(), annot = True)
```

```python
plt.figure(figsize=(35, 15))
plt.suptitle("Countplots of Numerical Columns")
for i in enumerate(df[['carat', 'depth', 'table', 'x', 'y', 'z', 'price']]):
    plt.subplot(4,2, i[0]+1)
    ax = sns.countplot(x = i[1], data = df)
    plt.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(25, 15))
plt.suptitle("Countplots of Categorical Columns")
for i in enumerate(df[categorical_columns]):
    plt.subplot(2,2, i[0]+1)
    ax = sns.countplot(x = i[1], data = df)
    plt.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(25, 15))
plt.suptitle("Pieplots of Categorical Columns")
for i in enumerate(df[categorical_columns]):
    plt.subplot(2,2, i[0]+1)
    df[i[1]].value_counts().plot.pie(autopct="%.1f%%", ylabel="", startangle=90)
plt.show()
```

```python
cut_map = {'Premium':4, 'Very Good':3, 'Ideal':5, 'Good':2, 'Fair':1}
```

```python
df['color'].unique()
```

```python
color_map = {'F':3, 'J':7, 'G':4, 'E':2, 'D':1, 'H':5, 'I':6}
```

```python
df['clarity'].unique()
```

```python
clarity_map = {'VS2':4, 'SI2':2, 'VS1':5, 'SI1':3, 'IF':8, 'VVS2':6, 'VVS1':7, 'I1':1}
```

#### **Apply Map Function**

```python
df['cut'] = df['cut'].map(cut_map)
```

```python
df['color'] = df['color'].map(color_map)
```

```python
df['clarity'] = df['clarity'].map(clarity_map)
```
